```python
import random

def create_bag(n, m):
    assert n >= m
    return [1] * m + [0] * (n - m)

create_bag(3, 1)
```

```python
TESTS = 100
K = 29
bag = create_bag(100, 75)

def mu(lst):
    return sum(lst) / len(lst)

def var(lst):
    mean = mu(lst)
    return mu([(i - mean)**2 for i in lst])

means = []
for i in range(1000):
    samples = [random.sample(bag, k=K)[-1] for i in range(TESTS)]
    means.append(mu(samples))

mean_mean = mu(means)
print(mean_mean)
print(var(means))
    
```

```python
import plotly.express as px
fig = px.histogram(means)
fig.show()
```

```python
TESTS = 10000
K = 4
n, m = 185, 53
p = m / n * (m - 1) / (n - 1) + (n - m) / n * (n - m - 1) / (n - 1)
theoretical_var = p * (1 - p)
bag = create_bag(n, m)

def random_var(bag):
    sample = random.sample(bag, k=K)
    return int(sample[0] == sample[-1])


samples = [random_var(bag) for i in range(TESTS)]


print(f"Theoretical Mean: {p}")
print(f"Theoretical Var: {theoretical_var}")
print(f"EXP Mean {mu(samples)}")
print(f"EXP Var {var(samples)}")
```
